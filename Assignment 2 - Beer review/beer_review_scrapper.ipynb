```python
import os
import sys

import pandas as pd
from datetime import datetime as dt
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException

import collections
import string
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import wordnet
from nltk import pos_tag

import scipy.stats as ss
import matplotlib.pyplot as plt
import math

import re
from nltk.corpus import stopwords

from sklearn.manifold import MDS

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
```

```python
path_to_driver = os.getcwd() + "\\chromedriver"
sys.path.insert(0,path_to_driver)
```

```python
wd = webdriver.Chrome('chromedriver',options=chrome_options)

url = "https://www.beeradvocate.com/beer/top-rated/"
```

```python
wd.get(url)
```

```python
table_el = wd.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[2]/div[2]/div/div/div[3]/div/div/div[2]/table/tbody')
action_tags_el = table_el.find_elements_by_xpath("//td[@class='hr_bottom_light']//a")
```

```python
beer_profiles = {}
for i in action_tags_el:
    beer_profiles[i.text] = i.get_attribute('href')
```

```python
beer_reviews = []
for k,v in beer_profiles.items():
    print(k, v)
    wd.get(v)
    review_el = wd.find_elements_by_id("rating_fullview_content_2")
    
    
    for i in review_el:
        rev_test = i.text.split('\n')
        beer_reviews.append((k, ' '.join(rev_test[5:-2]), rev_test[2]))
```

```python
len(beer_reviews)
```

```python
beer_reviews_df = pd.DataFrame(beer_reviews, columns=['product_name', 'product_review', 'user_rating'])
beer_reviews_df['user_rating'] = beer_reviews_df['user_rating'].apply(lambda x : float(x.split(' ')[0].split('/')[0]))
beer_reviews_df['product_review'] = beer_reviews_df['product_review'].apply(lambda x : x.strip())
beer_reviews_df['product_name'] = beer_reviews_df['product_name'].apply(lambda x : x.strip())
```

```python
beer_reviews_df.to_csv('beer_reviews.csv')
```
