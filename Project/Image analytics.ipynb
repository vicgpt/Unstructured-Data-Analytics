```python
import xlrd
from google.cloud import vision
import os
import pandas as pd

Application_Credentials = 'key.json'
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = Application_Credentials
client = vision.ImageAnnotatorClient()
image = vision.Image()
```

```python
#-------
def getfaceparams(image):
    client2 = vision.ImageAnnotatorClient()
    #image = vision.Image(content=content)
    response = client2.face_detection(image=image)
    faces = response.face_annotations

    # Names of likelihood from google.cloud.vision.enums
    likelihood_name = ('UNKNOWN', 'VERY_UNLIKELY', 'UNLIKELY', 'POSSIBLE',
                       'LIKELY', 'VERY_LIKELY')
    
    likelihood_name_2 = ['UNKNOWN', 'VERY_UNLIKELY', 'UNLIKELY', 'POSSIBLE',
                       'LIKELY', 'VERY_LIKELY']
    
    print('Faces:')

    #print(len(df))
    #print(len(faces))
    
    for face in faces:
        #print(face)
        #if(likelihood_name[face.anger_likelihood] in likelihood_name_2):
            
            #print('anger: {}'.format(likelihood_name[face.anger_likelihood]))
        anger = '{}'.format(likelihood_name[face.anger_likelihood])
            #print(anger_l)
        #else:
            #anger_l.append(['None'])
        
        
        print('joy: {}'.format(likelihood_name[face.joy_likelihood]))
        joy = '{}'.format(likelihood_name[face.joy_likelihood])
        
        
        print('surprise: {}'.format(likelihood_name[face.surprise_likelihood]))
        surprise = '{}'.format(likelihood_name[face.surprise_likelihood])
                          

        vertices = (['({},{})'.format(vertex.x, vertex.y)
                    for vertex in face.bounding_poly.vertices])

        print('face bounds: {}'.format(','.join(vertices)))
        vertices_list = '{}'.format(','.join(vertices))
        
        return [anger,joy,surprise,vertices_list]
    

```

```python
loc = ("nike_data_4oct.xls")
wb = xlrd.open_workbook(loc)
sheet = wb.sheet_by_index(0)
sheet.cell_value(0, 0)
df = pd.DataFrame()
anger_l = []
joy_l = []
surprise_l = []
bound_l = []

# loop through every url, retreive the image and send to google vision
for i in range(sheet.nrows):
    image_src_temp = sheet.cell_value(i, 0)
    image.source.image_uri = image_src_temp
    
    response = client.label_detection(image=image)
    
    labels = response.label_annotations
    l = []
    for label in labels:
        l.append(label.description)
    s = ' '.join(l)
    print("---------")
    print(s)
    df = df.append({'URL': image_src_temp, 'Labels': s}, ignore_index=True)
    
    list_output = getfaceparams(image=image)
    print(list_output)
    
    if(list_output is None):
        anger_l.append('None')
        joy_l.append('None')
        surprise_l.append('None')
    else :
        print(list_output[0])
        anger_l.append(list_output[0])
        joy_l.append(list_output[1])
        surprise_l.append(list_output[2])
        
        
    #------
    
    

df['Anger'] = anger_l
df['Joy'] = joy_l
df['Surprise'] = surprise_l

    
df.to_excel("NikeLabels.xlsx",index=False)

```
